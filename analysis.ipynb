```python
data =pd.read_csv('all_data.csv')
data.head()
```

```python
##finding  Nan

nan = data[data.isna().any(axis = 1)]
nan.head()


data = data.dropna(how = 'all')
data.head()
```

```python
## dropping Nan

data = data.dropna(how = 'all')
data.head()
```

```python
data = data.dropna(how='any')
data.head()
```

```python
data = data[data['Order Date'].str[0:2] !='Or']
```

```python
data['Quantity Ordered']= pd.to_numeric(data['Quantity Ordered'])
data['Price Each']= pd.to_numeric(data['Price Each'])
data.head()
```

```python
## ADD MONTH COLUMN
data['Month'] = data['Order Date'].str[0:2]
data['Month'] = data['Month'].astype('int32')
data.head()
```

```python
#ADD SALES COLUMN

data['Sales'] = data['Quantity Ordered'] * data['Price Each']
data.head()
```

```python
##use the .lambda method to split the purchase address and get city
def city(address):
    return address.split(',')[1]

def state(address):
    return address.split(',')[2].split(' ')[1]

data['City'] = data['Purchase Address'].apply(lambda x: f"{ city(x)} ({state(x)})")

data.head()
```

 Q1: WHAT WAS THE BEST MONTH FOR SALES AND HOW MUCH WAS EARNED?

```python
results = data.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt
```

```python
months = range(1,13)

plt.bar(months, results['Sales'])
plt.xticks(months)
plt.xlabel('month number')
plt.ylabel('Sales in USD($)')
plt.show()
```

Q2:WHAT CITY HAD THE HIGHEST NUMBER OF SALES

```python
results1 = data.groupby('City').sum()
results1
```

```python
Cities = [city for city, df in data.groupby('City')]

plt.bar(Cities, results1['Sales'])
plt.grid()
plt.xticks(Cities, rotation = 'vertical')
plt.xlabel('City Name')
plt.ylabel('Sales in $')
plt.show()
```

Q3: WHAT TIME SHOULD ADVERTS BE DISPLAYED TO MAXIMIZE LIKELIHOOD OF CUSTOMER BUYING THE PRODUCT

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
```

```python
data['Hour'] = data['Order Date'].dt.hour
data['Minute'] = data['Order Date'].dt.minute
data.head()
```

```python
hours = [hour for hour, df in data.groupby('Hour')]
plt.plot(hours,data.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel('hours')
plt.ylabel('number of orders')
plt.grid()
plt.show
```

Q4: WHAT PRODUCT SOLD THE MOST

```python
group = data.groupby('Product')
qtt = group.sum() ['Quantity Ordered']

products = [product for product, df in group]
plt.bar(products, qtt)

plt.xticks(products, rotation = 'vertical', size = 10)
plt.xlabel('Product Ordered')
plt.ylabel('Quantity Ordered')
plt.show()
```
